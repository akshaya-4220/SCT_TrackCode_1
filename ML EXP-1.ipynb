"Imorting required modules to read the data from file and to train the model"

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

" Reading dataset from file and assigning to a variable "

```python
dataset=pd.read_csv('/content/car data.csv')
```

"Printing the dataset"

```python
dataset
```

"Dropping the columns where the dataset contains string value"

```python
dataset=dataset.drop(['Car_Name','Fuel_Type','Seller_Type','Transmission'],axis=1)
```

"Printing the dataset after droping the columns"

```python
dataset
```

"Description about dataset"

```python
dataset.describe()
```

"Plotting the relation between Year and Owner Attributes"

```python
plt.scatter(dataset['Year'],dataset['Owner'])
plt.xlabel('Year')
plt.ylabel('Owner')
plt.title('Year vs Owner')
plt.show
```

"Plotting the relation between Selling price and kms Driven"

```python
plt.scatter(dataset['Selling_Price'],dataset['Kms_Driven'])
plt.xlabel('Selling_Price')
plt.ylabel('Kms_Driven')
plt.title('Selling_Price vs Kms_Driven')
plt.show
```

"Plotting the relation between Selling price and Present price"

```python
plt.scatter(dataset['Selling_Price'],dataset['Present_Price'])
plt.xlabel('Selling_Price')
plt.ylabel('Present_Price')
plt.title('Selling_Price vs Present_Price')
plt.show
```

"Finding the correlation coefficient"

```python
dataset.corr()
```

"Plotting heatmap for the dataset"

```python
sns.heatmap(dataset.corr(),annot=True)
```

"Plotting pairwise relationships in dataset"

```python
sns.pairplot(dataset)
```

"Separating the input attributes"

```python
x=dataset[['Selling_Price','Year','Kms_Driven','Owner']]
```

```python
x
```

```python
x.shape
```

"Separating class variables"

```python
y=dataset[['Present_Price']]
```

```python
y.shape
```

"Splitting the dataset into 4 parts such that the input and output variables are divided for
training and testing purpose by using train_test_split() method. Here the size of test data to be
splitted is defined by test_size and the remaining data is used for training"

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(x,y,test_size=0.2,random_state=10)
```

```python
X_train
```

```python
X_test
```

```python
Y_train
```

"Importing required modules"

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
```

"The model is trained by using linear regression algorithm"

```python
model=LinearRegression()
```

"Trains the model"

```python
model.fit(X_train,Y_train)
```

"Accessing the coefficient or weights of trained model"

```python
model.coef_
```

"Accessing the intercept value of a trained value"

```python
model.intercept_
```

"Overlaying a prediction line from the trained model"

```python
plt.scatter(X_train['Selling_Price'],Y_train)
plt.plot(X_train['Selling_Price'],model.predict(X_train),color='black')
plt.xlabel('Selling_Price')
plt.ylabel('Present_Price')
plt.title('Selling_Price vs Present_Price')
plt.show()
plt.scatter(X_train['Year'],Y_train)
plt.plot(X_train['Year'],model.predict(X_train),color='black')
plt.xlabel('Year')
plt.ylabel('Present_Price')
plt.title('Year vs Present_Price')
plt.show()
plt.scatter(X_train['Kms_Driven'],Y_train)
plt.plot(X_train['Kms_Driven'],model.predict(X_train),color='black')
plt.xlabel('Kms_Driven')
plt.ylabel('Present_Price')
plt.title('Kms_Driven vs Present_Price')
plt.show()
plt.scatter(X_train['Owner'],Y_train)
plt.plot(X_train['Owner'],model.predict(X_train),color='black')
plt.xlabel('Owner')
plt.ylabel('Present_Price')
plt.title('Owner vs Present_Price')
plt.show()
```

"Overlaying a prediction line from the tested model"

```python
plt.scatter(X_test['Selling_Price'],Y_test)
plt.plot(X_test['Selling_Price'],model.predict(X_test),color='black')
plt.xlabel('Selling_Price')
plt.ylabel('Present_Price')
plt.title('Selling_Price vs Present_Price')
plt.show()
plt.scatter(X_test['Year'],model.predict(X_test),color='black')
plt.xlabel('Year')
plt.ylabel('Present_Price')
plt.title('Year vs Present_Price')
plt.show()
plt.scatter(X_test['Kms_Driven'],Y_test)
plt.plot(X_train['Kms_Driven'],model.predict(X_train),color='black')
plt.xlabel('Kms_Driven')
plt.ylabel('Present_Price')
plt.title('Kms_Driven vs Present_Price')
plt.show()
plt.scatter(X_test['Owner'],Y_test)
plt.plot(X_train['Owner'],model.predict(X_train),color='black')
plt.xlabel('Owner')
plt.ylabel('Present_Price')
plt.title('Owner vs Present_Price')
plt.show()
```

"Checking whether the model is predicting correct values or not"

```python
model.predict([[7.25,2017,6900,0]])
```

"Testing the accuracy of model by r2 score method"

```python
score=r2_score(Y_test,model.predict(X_test))
score
```

"Testing error of the model by mean squared error method"

```python
mse=mean_squared_error(Y_test,model.predict(X_test))
mse
```

"Testing error by using mean absolute error method"

```python
mae=mean_absolute_error(Y_test,model.predict(X_test))
mae
```
